```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
%matplotlib inline
```

```python
# Read the 'Advertising.csv' dataset
data_filename = 'Advertising.csv'

# Read data file using pandas libraries
df = pd.read_csv(data_filename)
```

```python
# Take a quick look at the data
df.head()
```

```python
# Assign TV advertising as predictor variable/ design matrix 'X' 
X = df[["TV"]]

# Assign sales as response variable 'y'
y = df[["Sales"]]
```

```python
# Split the dataset in training and testing with 80% training set 
# & set random state to 310 for reproducable results 
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 310)
```

```python
# Run the cell below to plot the data
plt.figure(figsize=(7,5))
plt.scatter(X_train, y_train, edgecolor = 'k',label='training set',color = '#B9D6D6',s=45)
plt.scatter(X_test, y_test, edgecolor = 'k',label='test set',color = 'darkblue',s=50)
plt.xlabel('TV',fontsize=16)
plt.ylabel('Sales',fontsize=16)
plt.legend(fontsize=16)
```

 Now that the data is set up properly use the sklearn 'LinearRegression' class to fit on the training set.

```python
# declare a class instance of LinearRegression and assign it to model
model = LinearRegression()

# call the 'fit' method on the training data
model.fit(X_train, y_train)

# Predict on the test set using the 'predict' method
y_pred_test = model.predict(X_test)
```

```python
### edTest(test_mse) ###
# Compute the MSE with the predicted values and print it
mse = mean_squared_error(y_pred_test, y_test)
print(f'The test MSE is {mse}')
```

```python
# Make a plot of the data along with the predicted linear regression
plt.figure(figsize=(7,5))
plt.scatter(X_train, y_train, edgecolor = 'k',label='training set',color = '#B9D6D6',s=45)
plt.scatter(X_test, y_test, edgecolor = 'k',label='test set',color = 'darkblue',s=50)
plt.plot(X_test, y_pred_test, label='model',color = '#E44E47',lw=2)
plt.xlabel('TV',fontsize=16)
plt.ylabel('Sales',fontsize=16)
plt.legend(fontsize=16)
```
