```python
# import required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
%matplotlib inline
```

```python
# Read the 'Advertising.csv' dataset
data_filename = 'Advertising.csv'

# Read data file using pandas libraries
df = pd.read_csv(data_filename)
```

```python
# Take a quick look at the data
df.head()
```

```python
# Assign TV advertising as predictor variable 'x' and sales as response variable 'y'
x = df[['TV', 'Radio']]
y = df['sales']
assert len(x) == len(y)
x.shape, y.shape
```

```python
#Split the dataset in training and testing with 80% training set & set random state to 310 for reproducable results 
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8,random_state=310)
```

```python
# Plot the data
# plt.figure(figsize=(7,5))
# plt.scatter(x_train, y_train, edgecolor = 'k',label='training set',color = '#B9D6D6',s=45)
# plt.scatter(x_test, y_test, edgecolor = 'k',label='test set',color = 'darkblue',s=50)
# plt.xlabel('TV',fontsize=16)
# plt.ylabel('Sales',fontsize=16)
# plt.legend(fontsize=16)
```

```python
# Use the sklearn function 'LinearRegression' to fit on the training set
model = LinearRegression()
model.fit(x_train, y_train)

# Now predict on the test set
y_pred_test = model.predict(x_test)
```

```python
### edTest(test_mse) ###

# Now compute the MSE with the predicted values and print it
mse = mean_squared_error(y_test, y_pred_test)
print(f'The test MSE is {mse:0.2f}')
```
